```python
import pandas as pd
import csv
from tqdm import tqdm
from datetime import datetime
```

```python
with open('taskevents.csv', newline='') as f:
    reader = csv.reader(f)
    rawfile = list(reader)
```

```python
data = rawfile[1:5602]
len(data)
rawfile[0]
```

```python
data[0]
```

```python
for row in tqdm(data):
    if row[2]=="":
        row.append(10000000000000)
        continue
    date_dt1 = datetime.strptime(row[2], '%d/%m/%y')
    date_dt2 = datetime.strptime(row[3], '%d/%m/%y')
    row[2] =date_dt1
    row[3] =date_dt2
    day = date_dt1-date_dt2
    
    row.append(day.days)
```

```python
newdata = {}
for row in tqdm(data):
    if row[1] not in newdata.keys():
        newdata[row[1]] = []
```

```python
for k in newdata.keys():
    
    emails = 0
    calls = 0
    meetings = 0
    tasks = 0
    events = 0
    other = 0
    first_contact = 100000000000
    first_month = 0
    three_month = 0
    
    for row in data:
        if row[1] == k:
            if row[6] == "Email":
                emails += 1
            if row[6] == "Call":
                emails += 1
            if row[6] == "Meeting":
                emails += 1
            if row[6] == "Task":
                emails += 1
            if row[6] == "Event":
                emails += 1
            if row[6] == "Other":
                emails += 1
            if row[7] < first_contact:
                first_contact = row[7]
            if row[7] < 30:
                first_month += 1
            if row[7] < 90:
                three_month += 1
    total_contacts = emails+calls+meetings+tasks+events+other
    
    newdata[k] = [k,emails,calls,meetings,tasks,events,other,total_contacts,first_contact,first_month,three_month]
    
```

```python
colnames = ["id","emails","calls","meetings","tasks","events","other","total_contacts","first_contact","first_month","three_month"]
for k,v in newdata.items():
    newrow = k+", "+str(v).strip('[]')
    print(newrow)
```

```python

f = open('taskevents_writer.csv', 'w')

with f:
    writer = csv.writer(f)
    writer.writerow(colnames)
    for k,v in newdata.items():
        writer.writerow(v)
```
